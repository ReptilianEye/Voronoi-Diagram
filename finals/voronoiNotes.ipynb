```python
def generate_uniform_points(left, right, n=10 ** 5):
    """
    Funkcja generuje rwnomiernie n punktw na kwadratowym obszarze od left do right (jednakowo na osi y) o wsprzdnych rzeczywistych
    :param left:  lewy kraniec przedziau
    :param right: prawy kraniec przedziau
    :param n: ilo generowanych punktw
    :return: tablica punktw w postaci krotek wsprzdnych np. [(x1, y1), (x2, y2), ... (xn, yn)]
    """
    return [np.random.uniform(left, right, 2) for _ in range(n)]
```
